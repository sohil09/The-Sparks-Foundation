```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np
import matplotlib as mlt
import matplotlib.pyplot as plt
```

```python
# Reading data from Given Url
url="http://bit.ly/w-data"
df = pd.read_csv(url)
print("Data imported successfully")
df
```

```python
# Plotting the distribution of scores
df.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
# Ploting the Regression line
x=df.Hours
y=df.Scores
from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))



df.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')
plt.scatter(x,y)
plt.plot(x, mymodel)
plt.show()
print("r=",r)
```

```python
#Creating the Data for Calculating Errors
x_test=x[::5]
y_test=y[::5]
y_predict=myfunc(x_test)
df1=pd.DataFrame({"Actual" :y_test , "Predicted": y_predict })
df1
```

```python
#Calculation Of Erros
from sklearn import metrics
print("Mean Absolute Error :", metrics.mean_absolute_error(y_test,y_predict))
print("Mean Squared Error :", metrics.mean_squared_error(y_test,y_predict))
print("Root Mean Squared Error :", np.sqrt(metrics.mean_absolute_error(y_test,y_predict)))
```

**Predicted Score**

```python
hours=9.25
score=myfunc(hours)
print("No of Hours : {}".format(hours))
print("Predicted score : {}".format(score) )
```
