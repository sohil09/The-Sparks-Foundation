```python
####  Importing Libraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sea
```

```python
## reading the csv file
df=pd.read_csv(r'C:\Users\Harshit Boghara\Downloads\Iris.CSv')
df
```

```python
##removed the column "species" so that a graph can plotted
df.drop(["Species"], axis='columns' ,inplace=True)
df
```

```python
##  by using k-means and assuming 4 clusters every row is now a part of a cluster

from sklearn.cluster import KMeans

km = KMeans(n_clusters=4)
yp = km.fit_predict(df)
yp
```

```python
# adding clusters to datsframe
df['Cluster']= yp
df.head()
```

```python
# created 4 diffrent datasets to visualise data of tow columns
df1 = df[df.Cluster==0]
df2 = df[df.Cluster==1]
df3 = df[df.Cluster==2]
df4 = df[df.Cluster==3]
```

```python
plt.scatter(df1['SepalLengthCm'],df1['SepalWidthCm'],color='blue')
plt.scatter(df2['SepalLengthCm'],df2['SepalWidthCm'],color='green')
plt.scatter(df3['SepalLengthCm'],df3['SepalWidthCm'],color='yellow')
plt.scatter(df4['SepalLengthCm'],df4['SepalWidthCm'],color='pink')
```

```python
### created 4 diffrent datasets to visualise data of remaining two columns
plt.scatter(df1['PetalLengthCm'],df1['PetalWidthCm'],color='blue')
plt.scatter(df2['PetalLengthCm'],df2['PetalWidthCm'],color='green')
plt.scatter(df3['PetalLengthCm'],df3['PetalWidthCm'],color='yellow')
plt.scatter(df4['PetalLengthCm'],df4['PetalWidthCm'],color='pink')
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df)
    sse.append(km.inertia_)
sse    

# founded the sum of squared errors
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)

# plotting the elbow plot to determine the optimum value of k
```


From the elbow -plot it is clear that the optimum value of k=3
